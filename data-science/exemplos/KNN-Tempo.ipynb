```
# Definição das variáveis do problema (variáveis independentes) e das classes (variável dependente)
tempo = ['Ensolarado', 'Ensolarado', 'Nublado', 'Chuvoso', 'Chuvoso', 'Chuvoso', 'Nublado', 'Ensolarado', 'Ensolarado', 'Chuvoso', 'Ensolarado', 'Nublado', 'Nublado', 'Chuvoso']
temperatura = ['Quente', 'Quente', 'Quente', 'Suave', 'Frio', 'Frio', 'Frio', 'Suave', 'Frio', 'Suave', 'Suave', 'Suave' , 'Quente', 'Suave']
passeio = ['Não', 'Não', 'Sim', 'Sim', 'Sim', 'Não', 'Sim', 'Não', 'Sim', 'Sim', 'Sim', 'Sim', 'Sim', 'Não']
```

```
from sklearn import preprocessing # Biblioteca necessária para pré-processamento dos dados
```

```
le = preprocessing.LabelEncoder() # Classe para codificação dos dados nominais
```

```
tempo_codificada = le.fit_transform(tempo) # Codifica e normaliza os dados
```

```
print (tempo_codificada)
```

```
temperatura_codificada = le.fit_transform(temperatura)
classes = le.fit_transform(passeio)
```

```
caracteristicas = list(zip(tempo_codificada, temperatura_codificada)) # Une em tuplas as caracteríticas de cada objeto
print (caracteristicas)
```

```
from sklearn.neighbors import KNeighborsClassifier # Importação do KNN
```

```
modelo = KNeighborsClassifier(n_neighbors=5) # Criação do modelo de predição
```

```
modelo.fit(caracteristicas, classes) # Associa as amostras de treinamento com os rótulos das classes
```

```
print('Legenda')
dados = sorted(list(set(zip(tempo_codificada, tempo))))
print('Tempo: ', dados)
dados = sorted(list(set(zip(temperatura_codificada, temperatura))))
print('Temperatura: ', dados)
dados = sorted(list(set(zip(classes, passeio))))
print('Classes: ', dados)      
```

```
predicao = modelo.predict([[0,0]]) # Efetua a predição
print(predicao)
```
