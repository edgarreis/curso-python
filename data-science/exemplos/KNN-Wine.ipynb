```
from sklearn import datasets # Biblioteca de datasets
```

```
vinho = datasets.load_wine() # Escolha do dataset de vinhos
```

```
print (vinho.feature_names) # Mostra os rótulos das características
```

```
print (vinho.target_names) # Mostra os rótulos das classes
```

```
print (vinho.data[0:5]) # Mostra as 5 primeiras amostras do dataset
```

```
print(vinho.target) # Mostra os identificadores das classes
```

```
from itertools import groupby
[len(list(group)) for key, group in groupby(vinho.target)] # Agrupa os identificadores das classes e mostra a quantidade de amostras para cada uma delas
```

```
print(vinho.data.shape) # Mostra o tamanho total do dataset
```

```
from sklearn.model_selection import train_test_split # Importa a função responsável por fazer a divisão do dataset em amostras de treinamento e de teste
```

```
X_treino, X_teste, y_treino, y_teste = train_test_split(vinho.data, vinho.target, test_size=0.3) # Divide o dataset em 70% para treino e 30% para teste sem estratificação
```

```
print(y_treino) # Mostra os identificadores das classes das amostras selecionadas para treinamento
```

```
print(y_teste) # Mostra os identificadores das classes das amostras selecionadas para teste
```

```
import numpy
[len(list(group)) for key, group in groupby(numpy.sort(y_teste))] # Conta e mostra a quantidade de amostras de cada classe selecionadas para treinamento
```

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_treino, y_treino)
y_predito = knn.predict(X_teste)
print (y_predito)
acuracia = knn.score(X_teste, y_teste)
print (acuracia)
```

## Extra

```
from sklearn.model_selection import cross_val_score
import numpy as np


knn = KNeighborsClassifier(n_neighbors=3)
acuracias_vc = cross_val_score(knn, vinho.data, vinho.target, cv=5)


print(acuracias_vc)
print('Média das acurácias: {}'.format(np.mean(acuracias_vc)))
```
