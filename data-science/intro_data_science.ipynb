```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Lê o DataFrame
tmdb = pd.read_csv('tmdb_5000/tmdb_5000_movies.csv')

# Configura o tamanho da figura
plt.figure(figsize=(20, 8))

# Cria o gráfico de contagem das categorias em "original_language"
sns.countplot(x="original_language", data=tmdb)

# Ajusta a rotação dos rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=90)

# Mostra o gráfico
plt.show()
```

```python
# plt.pie(contagem_lingua["total"], labels = contagem_lingua["original_language"])
```

```python
total_por_lingua = tmdb["original_language"].value_counts()
total_de_ingles = total_por_lingua["en"]
total_geral = total_por_lingua.sum()
total_resto = total_geral - total_de_ingles
print(total_de_ingles, total_resto)
```

```python
# dataframe:  dicionário Py
dados = {
    'lingua' : ['ingles','outros'],
    'total' : [total_de_ingles, total_resto] 
}
dados = pd.DataFrame(dados)
```

```python
pd.DataFrame(dados)
```

```python
ax1 = sns.barplot(x="lingua", y="total", data=dados)

ax1.set_ylabel('Valor')

# Adiciona os valores acima das barras
for p in ax1.patches:
    ax1.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=8, color='black', xytext=(0, 5), textcoords='offset points')


```

```python
sns.barplot(x="lingua", y="total", data=dados)
```

```python
contagem_demais_lingua = tmdb.original_language.value_counts().to_frame().reset_index()
contagem_demais_lingua_sem_en = contagem_demais_lingua.drop(0)
contagem_demais_lingua_sem_en.columns = ['original language','total']
contagem_demais_lingua_sem_en.head()
```

```python
# Configura o tamanho da figura
plt.figure(figsize=(20, 8))

# Cria o gráfico de contagem das categorias em "contagem_demais_lingua_sem_en"
ax = sns.barplot(x='original language', y='total', data=contagem_demais_lingua_sem_en, palette='BuGn_d')
ax.set_ylabel('Valor')

# Adiciona os valores acima das barras
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=8, color='black', xytext=(0, 5), textcoords='offset points')



# Mostra o gráfico
plt.show()
```
